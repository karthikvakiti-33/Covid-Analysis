```python
columns_=['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']
df=df[columns_]
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=["continent"])
```

```python
df
```

```python
df=df.fillna(0)
df
```

```python
df.dtypes
```

```python
df["date"]=pd.to_datetime(df["date"])
```

```python
df["month"]=df["date"].dt.month
df
```

```python
df.dtypes
```

```python
df = df.astype({'total_cases': 'float64','total_deaths': 'float64','gdp_per_capita': 'float64','human_development_index':'float64'})
```

```python
df_groupby=df.groupby("continent").max()
df_groupby=df.reset_index()
df_groupby
```

```python
df_groupby["total_deaths_to_total_cases"]=df["total_deaths"]/df["total_cases"]
df_groupby
```

```python
import seaborn as sns
```

```python
sns.displot(df_groupby["gdp_per_capita"])
```

```python
sns.jointplot(data=df_groupby,x="total_cases",y="gdp_per_capita",kind="scatter")
```

```python
sns.pairplot(data=df_groupby)
```
